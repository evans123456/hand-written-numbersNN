```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd

mnist = tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
#plt.imshow(x_train[0])#shows with color
plt.imshow(x_train[0], cmap= plt.cm.binary) #shows blacknwhite
plt.show()
```

```python
# print(x_train[0])
pd.DataFrame(x_train[0]) #note how the matrix looks like a 5
```

```python
#normalizing -> scaling the values from 0 to 1 -> helps the model learn
x_train = tf.keras.utils.normalize(x_train,axis=1)
x_test = tf.keras.utils.normalize(x_test,axis=1)
# pd.DataFrame(x_train[0])
plt.imshow(x_train[0], cmap= plt.cm.binary) #shows blacknwhite
plt.show()
print(pd.DataFrame(x_train[0]))
```

```python
model = tf.keras.models.Sequential()#normal feed foward network

#creating the architecture of the model
#currently we have a 28x28 image and we want it flat
model.add(tf.keras.layers.Flatten())#input layer
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))#hidden layer(128 -> no of neurons,activation function)
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))#hidden layer
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))#output layer(softmax because its a probability distribution)

#parameters of the training of the model
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=['accuracy']) #optimizer= (where we can put stochastic gradient descent),loss is the degree of error (we can also use binary incase of cat vs dog ) ,metrics we want to track

#train the model
model.fit(x_train,y_train,epochs=3)
```

```python
#nn are good in fitting, but did they overfit?
#we calculate the validation loss and validation accuracy
val_loss,val_accuracy = model.evaluate(x_test,y_test)
print(val_loss,val_accuracy) #if there is a huge difference(these values and the train values(delta)) then you have probably overfit
```

```python
#save the model
model.save("num_reader.model")
```
